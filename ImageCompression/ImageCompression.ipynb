```python
from ImageCompression import ImageCompression
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("lena.jpg", 0)

image = cv2.resize(image, (image.shape[0] // 4, image.shape[1] // 4))
                         
compression = ImageCompression(image, 16, 8)
```

```python
plt.imshow(image, cmap="gray")
```

```python
decompressed_image = compression.decode(20)
```

```python
zero_image, zero_ssim = decompressed_image[0]
one_image, one_ssim = decompressed_image[1]
two_image, two_ssim = decompressed_image[2]
three_image, three_ssim = decompressed_image[3]
four_image, four_ssim = decompressed_image[4]
five_image, five_ssim = decompressed_image[5]
```

```python
plt.figure()

plt.subplot(2,3,1)
plt.imshow(zero_image, cmap="gray")
plt.title(f"0 iterations, SSIM: {round(zero_ssim, 2)}", fontsize=10)

plt.subplot(2,3,2)
plt.imshow(one_image, cmap="gray")
plt.title(f"1 iteration, SSIM: {round(one_ssim, 2)}", fontsize=10)

plt.subplot(2,3,3)
plt.imshow(two_image, cmap="gray")
plt.title(f"2 iterations, SSIM: {round(two_ssim, 2)}", fontsize=10)

plt.subplot(2,3,4)
plt.imshow(three_image, cmap="gray")
plt.title(f"3 iterations, SSIM: {round(three_ssim, 2)}", fontsize=10)

plt.subplot(2,3,5)
plt.imshow(four_image, cmap="gray")
plt.title(f"4 iterations, SSIM: {round(four_ssim, 2)}", fontsize=10)

plt.subplot(2,3,6)
plt.imshow(five_image, cmap="gray")
plt.title(f"5 iterations, SSIM: {round(five_ssim, 2)}", fontsize=10)


plt.tight_layout()
plt.show()
```

```python
decompressed_2 = ImageCompression(image, 2, 1)
print("2 complete")
decompressed_4 = ImageCompression(image, 4, 2)
print("4 complete")
decompressed_8 = ImageCompression(image, 8, 4)
print("8 complete")
decompressed_16 = ImageCompression(image, 16, 8)
print("16 complete")
decompressed_32 = ImageCompression(image, 32, 16)
print("32 complete")
decompressed_64 = ImageCompression(image, 64, 32)
print("64 complete")
```

```python
decompressed_2_decoded = decompressed_2.decode(5)[-1]
decompressed_4_decoded = decompressed_4.decode(5)[-1]
decompressed_8_decoded = decompressed_8.decode(5)[-1]
decompressed_16_decoded = decompressed_16.decode(5)[-1]
decompressed_32_decoded = decompressed_32.decode(5)[-1]
decompressed_64_decoded = decompressed_64.decode(5)[-1]
```

```python

plt.scatter([(128 / 2) ** 2, (128 / 8) ** 2, (128 / 16) ** 2, (128 / 32) ** 2, (128 / 64) ** 2], [decompressed_2_decoded[1], decompressed_8_decoded[1], decompressed_16_decoded[1], decompressed_32_decoded[1], decompressed_64_decoded[1]])
plt.plot([(128 / 2) ** 2, (128 / 8) ** 2, (128 / 16) ** 2, (128 / 32) ** 2, (128 / 64) ** 2], [decompressed_2_decoded[1], decompressed_8_decoded[1], decompressed_16_decoded[1], decompressed_32_decoded[1], decompressed_64_decoded[1]])
plt.xlabel("Number of Blocks")
plt.ylabel("SSIM Value")
plt.title("Number of blocks vs ssim")
```
